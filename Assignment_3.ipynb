# Question2:What do you think about the students' parental level of education?

# Question4:What do you think about the scoring variation for math, reading and writing based on
#● Gender
#● Test preparation course

```python
variation_scores_gender=students.groupby("gender")[['math score','math score','writing score']].std()
variation_scores_test_preparation_course=students.groupby("test preparation course")[['math score','math score','writing score']].std()
print("The Variation scores of students based on Gender:")
print(variation_scores_gender)
print("The Variation scores of students based on Test preparation course: ")
print(variation_scores_test_preparation_course)
```
